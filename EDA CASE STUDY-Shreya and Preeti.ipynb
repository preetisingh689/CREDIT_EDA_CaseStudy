```python
# method to get Meta-Data about any dataframe passed 
def get_data(dataframe) :
    data = pd.DataFrame({
            
                    'Total_Element': dataframe.count(), # total elements in columns
                    'Datatype' : dataframe.dtypes, # data types of columns
                    'Null_Count': dataframe.isnull().sum(), # total null values in columns
                    'Null_Percentage': dataframe.isnull().sum()/len(dataframe) * 100 # percentage of null values
        
                       })
    return data
```

```python
loan_data = get_data(df)
loan_data
```

```python
# listing the null values columns having more than 30%
high_null_count=loan_data[loan_data.Null_Percentage>=50]
high_null_count
```

```python
high_null_count.shape
```

### <font color=blue> Analysis -We can remove above 41 columns where null values are greater than or equal to 50%. </font>

```python
#Delete columns where missing values are more than or equal t0 50% .
drop_missing_value_col = high_null_count.index.tolist()
df.drop(drop_missing_value_col,inplace=True,axis=1)
```

```python
#checking the columns after removing high null values columns using user-defined function get_data
get_data(df)
```
